```python
# Define a new cgraph
cgraph = {'WA': ['SA', 'NT'],
          'Q': ['SA', 'NT', 'NSW'],
          'T': [],
          'V': ['SA', 'NSW'],
          'SA': ['WA', 'NT', 'Q', 'NSW','V'],
          'NT': ['Q', 'WA', 'SA'],
          'NSW': ['Q', 'V', 'SA']}
```

```python
import random

def get_color(node, rv, cg):
    # Check if no colors are available for the given node
    if not rv[node]:
        return 'no color'

    color = random.choice(rv[node])
    rv_copy = rv.copy()  # Create a copy to avoid modifying the original dictionary
    rv_copy.pop(node)

    # Update remaining values for neighboring nodes
    adjList = cg[node]
    for adj in adjList:
        if adj in rv_copy and color in rv_copy[adj]:
            rv_copy[adj].remove(color)

    return color
```

## Highest Degree Node

```python
# write a method to get the highest degree node
def get_hdnode(cg, solved):
    mnode = ''
    degree = -1
    for k in cg.keys():
        if len(cg[k])  > degree and k not in solved.keys():
            #print(k)
            degree = len(cg[k])
            mnode = k
    return mnode
```

## Minimum Remaining Value

```python
def get_MRVnode(rv):
    min_values = float('inf')
    min_node = None

    for node, values in rv.items():
        if len(values) < min_values:
            min_values = len(values)
            min_node = node

    return min_node
```

Color the nodes using MRV heuristic

```python
assigned = {}
remainingValues = {'WA': ['R', 'G', 'B'],
                    'Q': ['R', 'G', 'B'],
                    'T': ['R', 'G', 'B'],
                    'V': ['R', 'G', 'B'],
                    'NT': ['R', 'G', 'B'],
                    'SA': ['R', 'G', 'B'],
                    'NSW': ['R', 'G', 'B']}

for i in range(len(cgraph)):
    # assign color to the node with minimum remaining values (MRV)
    print("Remaining Values:", remainingValues)
    selectedNode = get_MRVnode(remainingValues)
    color = get_color(selectedNode, remainingValues, cgraph)

    # check if 'no color' is returned
    if color == 'no color':
        print("Failure: No color available for node", selectedNode)
        break

    assigned[selectedNode] = color
    print("Assigned:", assigned)
```

**Select node randomly**

```python
def get_RandomNode(rv):
    remaining_nodes = [k for k in rv.keys()]
    node = random.choice(remaining_nodes)
    return node
```
